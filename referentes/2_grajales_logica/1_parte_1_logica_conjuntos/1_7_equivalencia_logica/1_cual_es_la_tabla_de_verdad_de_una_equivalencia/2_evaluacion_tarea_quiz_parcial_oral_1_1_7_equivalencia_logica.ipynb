```python
# algoritmo para llamar a evaluación oral individual
import pandas as pd
import numpy as np

# Cargar la lista de estudiantes desde la URL
path = '/home/marco/Documentos/docencia/groups_list/fundamentos_aritmetica.csv'
estudiantes = pd.read_csv(path)
estudiantes['Parcial'] = estudiantes['Parcial'].astype(float) 

# Convertir la columna 'Nombre' a una lista de Python
lista_estudiantes = estudiantes['Nombre'].tolist()

# Función para llamar a un estudiante aleatoriamente y registrar su nota de seguimiento
def eval_oral():
    if lista_estudiantes:
        estudiante_aleatorio = np.random.choice(lista_estudiantes)
        print(f"Llamando a {estudiante_aleatorio}...")
        asistencia = input(f"¿Se expresó bien y correco {estudiante_aleatorio}? (s/n): ").strip().lower()
        
        if asistencia == 's':
            # Obtener el índice del estudiante en el DataFrame
            indice = estudiantes[estudiantes['Nombre'] == estudiante_aleatorio].index[0]
            # Incrementar su valor en la columna 'Parcial'
            estudiantes.at[indice, 'Parcial'] += 0.1
            print(f"Actualizado: {estudiante_aleatorio} ahora tiene {estudiantes.at[indice, 'Parcial']} en la columna 'Parcial'.")
        
        lista_estudiantes.remove(estudiante_aleatorio)
    else:
        print("Se han llamado a todos los estudiantes del grupo completo.")

# Guardar los cambios en el archivo CSV
def guardar_cambios():
    estudiantes.to_csv(path, index=False)
    print("Cambios guardados en el archivo CSV.")

# Ejemplo de uso
while lista_estudiantes:
    eval_oral()
    guardar_cambios()


```

### **Evaluación Oral: Lógica Matemática**

1. **Pregunta**: ¿Qué es una proposición lógica?
   **Respuesta**: Una proposición lógica es un enunciado que puede ser verdadero o falso, pero no ambas cosas simultáneamente.

2. **Pregunta**: Define la operación lógica de la conjunción ($p \land q$).
   **Respuesta**: La conjunción de dos proposiciones $p$ y $q$ es verdadera solo cuando ambas $p$ y $q$ son verdaderas; en cualquier otro caso, es falsa.

3. **Pregunta**: ¿Cuál es el valor de verdad de la disyunción ($p \lor q$) cuando $p$ es falsa y $q$ es verdadera?
   **Respuesta**: El valor de verdad de $p \lor q$ es verdadero, ya que al menos una de las proposiciones es verdadera.

4. **Pregunta**: Explica la diferencia entre una tautología y una contradicción.
   **Respuesta**: Una tautología es una proposición que es verdadera en todas las combinaciones posibles de sus variables, mientras que una contradicción es una proposición que es falsa en todas las combinaciones posibles.

5. **Pregunta**: ¿Qué significa que dos proposiciones sean lógicamente equivalentes?
   **Respuesta**: Dos proposiciones son lógicamente equivalentes si tienen el mismo valor de verdad en todas las posibles combinaciones de verdad de sus variables.

6. **Pregunta**: Construye la tabla de verdad para la negación de $p$ ($\neg p$).
   **Respuesta**: La tabla de verdad para $\neg p$ es:

   | $p$ | $\neg p$ |
   |-----|----------|
   |  V  |    F     |
   |  F  |    V     |

7. **Pregunta**: ¿Qué es una contingencia?
   **Respuesta**: Una contingencia es una proposición que es verdadera en algunas combinaciones de valores de verdad y falsa en otras.

8. **Pregunta**: Si $p$: "El número 2 es par" y $q$: "El número 2 es primo", ¿cuál es el valor de verdad de $p \land q$?
   **Respuesta**: El valor de verdad de $p \land q$ es verdadero, ya que ambas proposiciones son verdaderas.

9. **Pregunta**: ¿Cómo se niega una disyunción ($p \lor q$)?
   **Respuesta**: La negación de una disyunción se expresa como $\neg(p \lor q) = \neg p \land \neg q$.

10. **Pregunta**: Define la implicación lógica ($p \Rightarrow q$).
    **Respuesta**: La implicación $p \Rightarrow q$ es falsa solo cuando $p$ es verdadera y $q$ es falsa; en todos los demás casos, es verdadera.

11. **Pregunta**: Da un ejemplo de una proposición que sea una tautología.
    **Respuesta**: Un ejemplo de tautología es $p \lor \neg p$, ya que siempre es verdadera sin importar el valor de $p$.

12. **Pregunta**: ¿Cuál es el valor de verdad de la proposición $p \land \neg p$?
    **Respuesta**: La proposición $p \land \neg p$ es una contradicción y siempre es falsa.

13. **Pregunta**: ¿Qué significa que una proposición sea indeterminada?
    **Respuesta**: Una proposición indeterminada es aquella que no es ni una tautología ni una contradicción, es decir, es una contingencia.

14. **Pregunta**: Explica la ley de De Morgan aplicada a la conjunción.
    **Respuesta**: La ley de De Morgan aplicada a la conjunción dice que $\neg(p \land q) = \neg p \lor \neg q$.

15. **Pregunta**: ¿Cuál es el valor de verdad de $p \Rightarrow p$?
    **Respuesta**: El valor de verdad de $p \Rightarrow p$ es siempre verdadero, por lo tanto, es una tautología.

16. **Pregunta**: Si $p$: "El número 5 es impar", ¿cuál es el valor de verdad de $\neg p$?
    **Respuesta**: El valor de verdad de $\neg p$ es falso, ya que $p$ es verdadera.

17. **Pregunta**: ¿Cómo se relaciona una implicación lógica con su contrapositiva?
    **Respuesta**: La contrapositiva de una implicación $p \Rightarrow q$ es $\neg q \Rightarrow \neg p$, y ambas son lógicamente equivalentes.

18. **Pregunta**: ¿Qué es una equivalencia lógica?
    **Respuesta**: Una equivalencia lógica es una relación entre dos proposiciones que tienen el mismo valor de verdad en todas las combinaciones posibles de sus variables.

19. **Pregunta**: ¿Es $p \lor p$ equivalente a $p$? Explica.
    **Respuesta**: Sí, $p \lor p$ es equivalente a $p$, ya que en ambas proposiciones el valor de verdad es el mismo en cualquier combinación de $p$.

20. **Pregunta**: Construye la tabla de verdad para la implicación $p \Rightarrow \neg q$.
    **Respuesta**:

    | $p$ | $q$ | $\neg q$ | $p \Rightarrow \neg q$ |
    |-----|-----|----------|------------------------|
    |  V  |  V  |    F     |           F            |
    |  V  |  F  |    V     |           V            |
    |  F  |  V  |    F     |           V            |
    |  F  |  F  |    V     |           V            |

21. **Pregunta**: ¿Cuál es la implicación contraria de $p \Rightarrow q$?
    **Respuesta**: La contraria de $p \Rightarrow q$ es $q \Rightarrow p$, que no necesariamente es lógicamente equivalente a la implicación original.

22. **Pregunta**: Define la bicondicional ($p \Leftrightarrow q$).
    **Respuesta**: La bicondicional $p \Leftrightarrow q$ es verdadera solo cuando ambas $p$ y $q$ son verdaderas o ambas son falsas; en cualquier otro caso, es falsa.

23. **Pregunta**: Si $p$: "El número 10 es divisible por 5" y $q$: "El número 10 es impar", ¿cuál es el valor de verdad de $p \Rightarrow q$?
    **Respuesta**: El valor de verdad de $p \Rightarrow q$ es falso, ya que $p$ es verdadera y $q$ es falsa.

24. **Pregunta**: ¿Es $\neg(p \land q)$ equivalente a $\neg p \lor \neg q$?
    **Respuesta**: Sí, $\neg(p \land q)$ es equivalente a $\neg p \lor \neg q$, según la ley de De Morgan.

25. **Pregunta**: ¿Qué representa la operación $p \oplus q$ (disyunción exclusiva)?
    **Respuesta**: La operación $p \oplus q$ es verdadera solo cuando una de las proposiciones es verdadera y la otra es falsa; si ambas son verdaderas o ambas son falsas, la disyunción exclusiva es falsa.

# Autoevaluación 

```python
import csv

def recolectar_datos_autoavaluacion():
    # Definir el nombre del archivo CSV
    nombre_archivo = 'autoevaluacion_estudiantes.csv'
    
    # Definir los encabezados para el archivo CSV
    encabezados = ['Nombre', 'Fortalezas', 'Áreas de Mejora']
    
    # Lista de posibles fortalezas para que el estudiante pueda elegir
    posibles_fortalezas = [
        "Resolución de problemas",
        "Pensamiento crítico",
        "Razonamiento lógico",
        "Capacidad de análisis",
        "Trabajo en equipo",
        "Comunicación efectiva",
        "Dominio de conceptos matemáticos",
        "Manejo de herramientas tecnológicas"
    ]
    
    # Mostrar la lista de fortalezas y permitir que el estudiante elija
    print("Seleccione sus fortalezas de la siguiente lista (puede elegir varias separadas por comas):")
    for i, fortaleza in enumerate(posibles_fortalezas, 1):
        print(f"{i}. {fortaleza}")
    
    seleccion = input("Ingrese los números correspondientes a sus fortalezas (ejemplo: 1, 3, 5): ")
    
    # Convertir la selección en una lista de fortalezas
    seleccion_fortalezas = [posibles_fortalezas[int(i) - 1] for i in seleccion.split(',') if i.isdigit() and 1 <= int(i) <= len(posibles_fortalezas)]
    
    # Solicitar las áreas de mejora
    areas_mejora = input("Describa las áreas en las que necesita mejorar: ")
    
    # Crear una lista con los datos del estudiante
    datos = [input("Ingrese su nombre: "), ', '.join(seleccion_fortalezas), areas_mejora]
    
    # Abrir el archivo CSV para agregar los datos
    with open(nombre_archivo, mode='a', newline='') as archivo_csv:
        escritor_csv = csv.writer(archivo_csv)
        
        # Verificar si el archivo está vacío para escribir los encabezados
        archivo_csv.seek(0, 2)  # Mover el cursor al final del archivo
        if archivo_csv.tell() == 0:  # Verificar si el archivo está vacío
            escritor_csv.writerow(encabezados)
        
        # Escribir los datos en el archivo CSV
        escritor_csv.writerow(datos)
    
    print("La autoevaluación se ha guardado correctamente.")

# Ejecutar la función para recolectar datos
recolectar_datos_autoavaluacion()
```
